## 1. Kamatszámítás (számtani és mértani haladvány)

Alaptőke: $N$.   
Kamatozási időszak: $T$.  
Kamatláb: $l$.  
Egyszerű kamat: $X_0=N$ és $X_n-X_{n-1}=l\cdot X_0$, minden $n\in\{1,2,\dots,T\}$ esetén.   
Kamatos kamat: $Y_0=N$ és $Y_n-Y_{n-1}=l\cdot Y_{n-1}$, minden $n\in\{1,2,\dots,T\}$ esetén. 

- Az **alaptőke** az a pénzösszeg, ami kezdetben a rendelkezésre áll. 
- A **kamatozási időszak** megadja, hogy hány évig kamatoztatjuk a rendelkezésre álló pénzt. 
- A **kamatláb** egy arány, ami alapján a kamatot számoljuk.

- Az **egyszerű kamatot** az alaptőke és a kamatláb szorzataként számoljuk.  
Ha $X_n$ jelöli a rendelkezésre álló pénzösszeget, akkor az $n.$ évben a kamat $l\cdot X_0$, ahonnan 
$$X_0=N,\quad X_n-X_{n-1}=l\cdot X_0,\quad \forall n=\overline{1,T}.$$  
A rekurzív egyenletet megoldva kapjuk, hogy 
$$X_n=N\cdot(1+ l\cdot n),\quad \forall n=\overline{1,T}.$$
Vegyük észre, hogy $X_n$ egy számtani haladvány (az egymásutáni tagok különbsége állandó).

- A **kamatos kamatot** az aktuálisan rendelkezésre álló pénzösszeg és a kamatláb szorzataként számoljuk.  
Ha $Y_n$ jelöli a rendelkezésre álló pénzösszeget, akkor az $n.$ évben a kamat $l\cdot Y_{n-1}$, ahonnan 
$$Y_0=N,\quad Y_n-Y_{n-1}=l\cdot Y_{n-1},\quad \forall n=\overline{1,T}.$$
A rekurzív egyenletet megoldva kapjuk, hogy 
$$Y_n=N\cdot(1+l)^n,\quad \forall n=\overline{1,T}.$$
Vegyük észre, hogy $Y_n$ egy mértani haladvány (az egymásutáni tagok aránya állandó).

## 2. Populációdinamika

Külöböző populációk egyedszámának időbeli vizsgálata elsőrendű rekurzióra vezetődik vissza a következő modellek alapján.

#### 2.1. Exponenciális modell
Ha $X_n$ jelöli egy populáció egyedeinek a számát az $n$-edik időpillanatban, akkor a modellt megadó rekurzió a következő:
$$X_{n+1}-X_n=r\cdot X_n,\tag{1}$$
vagyis az időegységenkénti új egyedek száma arányos a meglévő egyedek számával (természetes szaporulat).

#### 2.2. Logisztikus modell
Ha $Y_n$ jelöli egy populáció egyedeinek a számát az $n$-edik időpillanatban, akkor a modellt megadó rekurzió a következő:
$$Y_{n+1}-Y_n=r\cdot Y_n \left(1-\frac{Y_n}{M}\right),\tag{2}$$
vagyis az időegységenkénti új egyedek száma arányos a meglévő egyedek számával, illetve a még elférő egyedek számával.


**Magyarázat.**
- Az **exponenciális modell** esetében feltételezzük, hogy a populáció élettere ideális, azaz semmi nem gátolja az egyedek szaporodását.
- A **logisztikus modell** esetén feltételezzük, hogy a populáció életterének befogadóképessége véges, vagyis az adott életéren legfeljebb $M$ darab egyed fér el. 

**Példa.**
- Ha tekintünk egy baktérium populációt, amelyik szabadon szaporodik, akkor az egyedek számát az exponenciális modell adja meg.
- Ha az előbbi populáció életterét korlátozzuk (például bezárjuk őket egy befőttes üvegbe), akkor az egyedek számát a logisztikus modell adja meg.

## 3. Diszkrét logisztikus egyenlet

A **diszkrét logisztikus egyenlet** a következő:    
$$U_{n+1}=k\cdot U_n \cdot (1-U_n),\tag{3}$$
ahol $k\in[0,4]$ és  $U_n\in(0,1)$, minden $n\in\mathbb{N}$ esetén.

Lásd logisztikus modell ($k=r+1$, $U_n=\frac{r}{k M} Y_n$.)

**Magyarázat.** A (2)-es egyenlettel megadott logisztikus modell, igen hasznos, mivel számos jelenségre ad magyarázatot.  
A probléma az, hogy tetszőleges $r$ és $M$ paraméterek esetén az egyenletet teljesítő sorozatok általános tagja nem meghatározható.  
A sorozatok természetét viszont általános esetben is tudjuk vizsgálni.  
Első lépésként egyszerűbb alakra hozzuk a (2)-es egyenletet a következőképpen. Az
$$Y_{n+1}-Y_n=r\cdot Y_n \left(1-\frac{Y_n}{M}\right).$$
egyenletet átírható, mint 
$$Y_{n+1}=Y_n\left(1+r-\frac{rY_n}{M}\right)=(1+r)Y_n\left(1-\frac{rY_n}{(r+1)M}\right).$$
Legyen $k=r+1$. Beszorozva mindkét oldalt $\frac{r}{k M}$-el azt kapjuk, hogy
$$\frac{r}{k M} Y_{n+1}=k \frac{r}{k M}Y_n\left(1-\frac{r}{k M}\cdot  Y_n\right).$$
Most legyen $U_n=\frac{r}{k M} Y_n$, ekkor kapjuk, hogy
$$U_{n+1}=k\cdot U_n \cdot (1-U_n).$$

**Megjegyzés.** Vegyük észre, hogy $U_n$ arányos a populáció telítettségével (az aktuális egydszám és a maximális egyedszám arányával), vagyis $\frac{Y_n}{M}$-el.

Vegyük észre, hogy különböző $k$ értékek esetén a sorozat különbözőképpen viselkedik:

- a $k=3{,}3$ esetben a sorozat értékei (egy adott index után) két érték között váltakoznak, a sorozat ilyenkor **periódikus**, és a periódusa $2$; 
- a $k=3{,}5$ esetben a sorozat szintén periódikus, ebben az esetben viszont a periódusa $4$;
- a $k=3{,}7$ esetben a sorozat **kaotikus**.

**Példa.** 
- Konvergens sorozat például az $x_n=\frac{1}{n}$. Ebben az esetben $\lim_{n\to\infty} x_n=0$.
- 2-periódiukus sorozat például az $x_n=(-1)^n$. Ebben az esetben $x_{2k}=1$ és $x_{2k+1}=-1$, minden $k\in\mathbb{N}$ esetén.
- Kaotikus sorozat például, a $\sin(n)$, ahol $n\in\mathbb{N}$.

Készítsd el a diszkrét logisztikus egyenlet bifurkációs digrammját! 

**Magyarázat.** A bifurkációs diagramról leolvasható, hogy adott $k$ esetén, milyen az $U_n$ sorozat természete.   
Adott $k$ esetén megnézzük, hogy a $(k,0)$ ponton keresztül húzott függőleges egyenes hány pontban metszi a grafikont.
- Ha egy pontban metszi, akkor azt jelenti, hogy a sorozat tagjai egy adott, index után egyenlőek lesznek, vagyis a sorozat **konvergens**.
- Ha két, három, négy stb. pontban metszi, akkor azt jelenti, hogy a sorozat tagjai ezen érték között váltakoznak. Ide tartozik az az eset is, amikor a sorozat 2-, 3-, 4-**periodikus**.

**Példák.** 
- $k\in(0,1)$ esetén a sorozat konvergens és $0$-hoz tart.
- $k\in[1,3)$ esetén a sorozat konvergens és $\frac{k-1}{k}\in[0,1)$-hez tart.
- $k\in(3,1+\sqrt{6}\approx3{,}449)$ esetén a sorozat tagjai két érték között váltakoznak.

### 4.2. Pókháló diagram

Egy másik mód a rekurzív sorozatok viselkedésének tanulmányozására az úgynevezett pókhálódiagram.  
Például $k=2{,}7$ esetén az $U_0=0{,}5$ és $U_{n+1}=k\cdot U_n\cdot(1-U_n)$ sorozathoz rendelt pókhálódiagram a következő.

![cob1.png]( )

Az ábrát a következő lépésekben szerkesztettük:
- kék színnel berajzoltuk az $f\colon[0,1]\to\mathbb{R}$, $f(x)=k\cdot x \cdot(1-x)$ függvényt (ez adja meg, hogy egy tagból, hogyan kapjuk a következőt);
- pirossal berajzoltuk az első szögfelezőt ($y=x$ egyenletű egyenes);
- feketével pedig egy törött vonalat rajzoltunk, amelyet úgy kaptunk, hogy:
  - kiindultunk az $(U_0,0)$ koordinátájú pontból;
  - húztunk egy függőleges szakaszt, amíg el nem metszettük metszettük az $f$ grafikonját;
  - a metszéspontból húztunk egy vízszintes szakaszt, amíg el nem metszettük a szögfelezőt;
  - az utolsó két lépést ismételtük $100$-szor.

**Megjegyzések.** 
- Azok az $(x,y)$ pontok, amelyek az $f$ függvény grafikonjának és a szögfelelezőnek a metszéspontjai, teljesül, hogy $x=f(x)$.  
  Ezeket az $x$ értékeket, az $f$ függvény **fixpontjainak** nevezzük.    
- A fekete törött vonalat meghatározó pontok koordinátái rendre:
$$(U_0,0);\ \  (U_0,U_1);\ \ (U_1,U_1);\ \ (U_1,U_2);\ \ (U_2,U_2);\ \ \dots $$
- Ez alapján a fekete törött vonal, illetve a függvénygrafikon metszéspontjainak abszcisszái ($x$-koordinátái) pontosan az $U_n$ sorozat tagjai.


A pókháló diagram segítségével következtethetünk a sorozat természetére:

|  |  |
|--|--|
|konvergens|2-periodikus|
|4-periodikus|kaotikus|

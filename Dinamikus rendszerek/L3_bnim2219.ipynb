<h3 style="background :Red; padding: 1%!important;"> 3. Házi feladat. (+2 pont)</h3>

Az alábbi ábrán látható Sierpiński gúla alapja egy $2$ oldalhosszúságú négyzet, magassága $1$.</br> Generáld ki a gúlát $10000$ iterációval az origóból indulva és írasd ki a megfelelő transzformációs táblázatot! </br> **Fontos:** A kapott gúla mérete és pozíciója **pontosan** kell egyezzen az alábbi ábrán láthatóéval!

```python
colors = ['b', 'r', 'g', 'c', 'm', 'y', 'k']

def fractal(data, ax):
    point = [[0], [0], [0]]
    pointsX = []
    pointsY = []
    pointsZ = []
    probabilities = []
    indexes = []
    for i in range(len(data)):
        indexes.append(i)
        probabilities.append(data[i][12])
        pointsX.append([])
        pointsY.append([])
        pointsZ.append([])
    for i in range(10000):
        #choosing random row from input
        indexF = np.random.choice(indexes, p=probabilities)
        scaleMatrix = []
        translation = []
        index = 0
        #scaling matrix
        for j in range(3):
            scaleMatrix.append([])
            for k in range(3):
                scaleMatrix[j].append(data[indexF][index])
                index += 1
        #translation vector
        for j in range(3):
            translation.append([data[indexF][index]])
            index += 1
        point = np.matmul(scaleMatrix, point)
        point = np.add(point, translation)
        pointsX[indexF].append(point[0][0])
        pointsY[indexF].append(point[1][0])
        pointsZ[indexF].append(point[2][0])
    for i in range(len(pointsX)):
        ax.scatter3D(pointsX[i], pointsY[i], pointsZ[i], c=colors[i], marker='.')

serp_pyramid = [[1/2, 0, 0, 0, 1/2, 0, 0, 0, 1/2, -1/2, -1/2, 0, 1/5],
               [1/2, 0, 0, 0, 1/2, 0, 0, 0, 1/2, -1/2, 1/2, 0, 1/5],
               [1/2, 0, 0, 0, 1/2, 0, 0, 0, 1/2, 1/2, -1/2, 0, 1/5],
               [1/2, 0, 0, 0, 1/2, 0, 0, 0, 1/2, 1/2, 1/2, 0, 1/5],
               [1/2, 0, 0, 0, 1/2, 0, 0, 0, 1/2, 0, 0, 1/2, 1/5],
               ]

fig7 = plt.figure()
ax7 = plt.axes(projection='3d')
fractal(serp_pyramid, ax7)
```

<h3 style="background :Red; padding: 1%!important;"> 4. Házi feladat. (4 pont)</h3>

Generáld ki az alábbi két táblázattal megadott Apollóniusz háromszöget, illetve korongot $10000$ iterációval az origóból indulva!

```python
ap_tri_data=[[1+np.sqrt(3),2+np.sqrt(3),-2-np.sqrt(3),5+3*np.sqrt(3),1/3],
    [(-1+1j*np.sqrt(3))*(1+np.sqrt(3)),(-1+1j*np.sqrt(3))*(2+np.sqrt(3)),-4-2*np.sqrt(3),10+6*np.sqrt(3),1/3],
    [(-1-1j*np.sqrt(3))*(1+np.sqrt(3)),(-1-1j*np.sqrt(3))*(2+np.sqrt(3)),-4-2*np.sqrt(3),10+6*np.sqrt(3),1/3]]

ap_circ_data=[[2.73,3.73,-3.73,10.2,1/3],
    [3.73-6.46j,-10.2+17.66j,5.46,7.46,1/3],
    [3.73+6.46j,-10.2-17.66j,5.46,7.46,1/3]]

colors = ['b.', 'r.', 'g.', 'c.', 'm.', 'y.', 'k.', 'C1.']

def fractal(data, ax):
    point = complex(0, 0)
    pointsX = []
    pointsY = []
    probabilities = []
    indexes = []
    for i in range(len(data)):
        indexes.append(i)
        probabilities.append(data[i][4])
        pointsX.append([])
        pointsY.append([])
    for i in range(10000):
        #choosing random row from input
        indexF = np.random.choice(indexes, p=probabilities)
        alfa = data[indexF][0]
        beta = data[indexF][1]
        gamma = data[indexF][2]
        delta = data[indexF][3]
        numerator = alfa * point + beta
        denominator = gamma * point + delta
        point = numerator / denominator
        pointsX[indexF].append(point.real)
        pointsY[indexF].append(point.imag)
    for i in range(len(pointsX)):
        ax.plot(pointsX[i], pointsY[i], colors[i])

        
fig8, ax8 = plt.subplots()
fractal(ap_tri_data, ax8)

fig9, ax9 = plt.subplots()
fractal(ap_circ_data, ax9)
```

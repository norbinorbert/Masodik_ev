### 1.3. Ciklusok és elágazások: ```for```, ```while```, ```if```

````python
for i in range(10):                      # teljes szintaxis: range(from=0,to,step=1) lásd később
    print(i, end=' ')                    # blokkok indentálással (4 szóköz)
````

````python
i = 0
while (i<10):
    if (i%3==0):
        print(i,end=' ')
    elif (i%3==1):
        print(-i,end=' ')
    else:
        print(0,end=' ')
    i = i+1
````

#### 1.3.1. Listák bejárása

````python
l1 = [1,3,5,6]
for i in l1:
    print(i, end=' ')
````

````python
for (i,val) in enumerate(l1):
    print(i,val,sep='-',end=' ')
````

````python
t1 = (2,4,6,8,10)
for (v1,v2) in zip(l1,t1):
    print(v1,v2,sep='-',end=' ')
````

#### 1.3.2 Listák generálása

````python
l = []
for i in range(0,10,2):
    l.append(i**2)
print(l)
````

````python
l = [i**2 for i in range(0,10)]           #list comprehesion
print(l)
````

````python
l = [i**2 for i in range(0,10) if i%2==0]
print(l)
````

## 3. Jupyter

Tutoriál: https://code.visualstudio.com/docs/datascience/jupyter-notebooks  
Interact & Widgets: https://ipywidgets.readthedocs.io/en/stable/

````python
import ipywidgets as widgets

def update1(x):
    return x**2

widgets.interact(update1,x=(-5,5));
````

````python
plt.ion()       
import numpy as np
fig4,ax4 = plt.subplots()
x = np.linspace(0,np.pi*4,201)
def update2(k=1,l=0):
    ax4.cla()
    ax4.plot(x,np.sin(k*x)+l,'r-')
widgets.interact(update2,k=(0,2,0.1),l=(0,2,0.1))
````

````python
widgets.interact(lambda x: x, x=widgets.Play(0,0,100))
````

````python
fig5,ax5 = plt.subplots()
tt = np.linspace(0,2*np.pi,201)
def update3(t=0):
    t=t/100*2*np.pi
    ax5.cla()
    print(t)
    ax5.set_aspect('equal')
    ax5.plot(np.cos(tt),np.sin(tt))
    ax5.plot(np.cos(t),np.sin(t),'bo')
widgets.interact(update3,t=widgets.Play(0,0,100))
````

<h3 style="background :Red; padding: 1%!important;">4. Házi feladat. (3 pont)</h3>

Készíts egy ábrát, amelyre a ```plot``` függvény segítségével $n+1$ darab piros (körvonalú) négyzetet rajzolsz ki úgy, hogy azok oldalai párhuzamosak legyenek az $O_x$ és $O_y$ tengelyekkel, bal alsó sarkuk legyen az origó és amelyek oldalai rendre $2^0, 2^1, \dots, 2^n$ hosszúságúak! Ugyanerre az ábrára a ```plot``` függvény egyszeri meghívásával rajzolj egy-egy kék pontot a négyzetek jobb felső sarkaira! Az $n$ értéke interaktívan legyen változtatható egyesével a $0$ és $5$  értékek között, és alapértelmezetten legyen $3$.

<h3 style="background :Red; padding: 1%!important;">5. Házi feladat. (+2 pont)</h3>

Egy $(0,0)$ pozícióból $\varphi$ szög alatt $v_0$ kezdősebességgel elhajított test helyzetének koordinátáit a $t$ időpillanatban az

$$
\begin{cases}
x(t)=v_0\cos(\varphi)t\\
y(t)=v_0\sin(\varphi)t-5t^2
\end{cases}
$$

rendszer adja. Készíts egy animációt, amelyen megjelenik a $t\in[0,4]$ időintervallumnak megfelelő pálya, illetve ahol egy golyó az idő szerinti animációnak megfelelően végigfut az adott pályán! Az ismeretlen paraméterek interaktívan legyenek állíthatóak!
